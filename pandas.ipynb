```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("arianazmoudeh/airbnbopendata")

print("Path to dataset files:", path)
```

```python
import numpy as np
import pandas as pd
import os

print(os.listdir(path))
fil=os.path.join(path,"Airbnb_Open_Data.csv")
print(fil)
```

```python
data_set=pd.read_csv("/root/.cache/kagglehub/datasets/arianazmoudeh/airbnbopendata/versions/1/Airbnb_Open_Data.csv")
print(data_set.head(5))
```

```python
data_set=data_set.drop(["last review","reviews per month","license","lat","long"],axis=1)
null_data=data_set.isnull().sum()
```

```python
null_data=data_set.isnull().sum()
data_set["house_rules"].fillna("no rules",inplace=True)
print(null_data)
```

```python
data_set.dropna(axis=0,inplace=True)
print(data_set)
```

```python
data_set.info()
```

```python
data_set.drop_duplicates(inplace=True)
data_set.duplicated().sum()
```

```python
data_set.rename(columns={"level_0":"index"},inplace=True)
print(data_set.head(20))
```

```python
data_set['host_identity_verified']= data_set['host_identity_verified'].str.upper()
```

```python
data_set["price"]=data_set["price"].str.replace("$","")
data_set["price"]=data_set["price"].str.replace(",","")
data_set["price"]=data_set["price"].str.replace(" ","")
data_set["service fee"]=data_set["service fee"].str.replace("$","")
data_set["service fee"]=data_set["service fee"].str.replace(",","")
data_set["service fee"]=data_set["service fee"].str.replace(" ","")
```

```python
data_set["price"]=data_set["price"].astype(int)
data_set["service fee"]=data_set["service fee"].astype(int)
```
